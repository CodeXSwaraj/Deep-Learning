```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

```

```python
# Step 2: Load dataset
iris = load_iris()
X = iris.data
y = iris.target


```

```python
# Step 3: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Step 4: Applying PCA without specifying number of components to see explained variance
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

```

```python
# Step 5: Plot explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Components')
plt.show()

# Step 6: Choose number of components based on explained variance ratio
# You can select the number of components based on the plot above
# For instance, if you want to keep 95% of variance, find the number of components that gives you that.
# In this example, let's say you choose 2 components.

```

```python
# Step 7: Applying PCA with selected number of components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

```

```python
# Step 8: Visualization before and after PCA
plt.figure(figsize=(12, 6))

# Before PCA
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Before PCA')

```

```python
# Step 8: Visualization before and after PCA
plt.figure(figsize=(12, 6))

# After PCA
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('After PCA')

plt.tight_layout()
plt.show()
```
