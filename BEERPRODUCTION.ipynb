```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Read CSV file
df = pd.read_csv('beerpro.csv')

# Convert 'Month' column to datetime format
df['Month'] = pd.to_datetime(df['Month'])
```

```python
# Extract year and month
df['Year'] = df['Month'].dt.year
df['Month_Name'] = df['Month'].dt.strftime('%B')  # Full month name
```

```python
# Statistical description
print(df.describe())
```

```python
# Yearly boxplot
plt.figure(figsize=(14, 8))  # Set figure size
sns.boxplot(x='Year', y='Monthly beer production', data=df)  # Create boxplot
plt.title('Yearly Boxplot of Monthly Beer Production', fontsize=16)  # Set title
plt.xlabel('Year', fontsize=14)  # Set x-axis label
plt.ylabel('Monthly Beer Production', fontsize=14)  # Set y-axis label
plt.xticks(fontsize=12, rotation=45)  # Set x-tick labels
plt.yticks(fontsize=12)  # Set y-tick labels
plt.show()  # Show plot
```

yearly the sale is booming, 1976 in a month the sale was very high, in 1956 is the lowest sale.

```python
plt.figure(figsize=(14, 8))  # Set figure size
sns.lineplot(x='Year', y='Monthly beer production', data=df, estimator='mean', ci=None)  # Create line plot
plt.title('Trend of Monthly Beer Production Over the Years', fontsize=16)  # Set title
plt.xlabel('Year', fontsize=14)  # Set x-axis label
plt.ylabel('Monthly Beer Production', fontsize=14)  # Set y-axis label
plt.xticks(fontsize=12, rotation=45)  # Set x-tick labels
plt.yticks(fontsize=12)  # Set y-tick labels
plt.grid(True)  # Add grid
plt.show()  # Show plot
```

```python
# Monthly boxplot
plt.figure(figsize=(14, 8))  # Set figure size
sns.boxplot(x='Month_Name', y='Monthly beer production', data=df, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])  # Create boxplot
plt.title('Monthly Boxplot of Monthly Beer Production', fontsize=16)  # Set title
plt.xlabel('Month', fontsize=14)  # Set x-axis label
plt.ylabel('Monthly Beer Production', fontsize=14)  # Set y-axis label
plt.xticks(fontsize=12, rotation=45)  # Set x-tick labels
plt.yticks(fontsize=12)  # Set y-tick labels
plt.show()  # Show plot
```

PRODUCTION OF BEER IS DOWN IN JUNE MONTH, HIGHEST LEVEL OF PRODUCTION ARE IN DECEMBER AND JANUARY

```python
plt.figure(figsize=(55, 20))  # Set larger figure size for better readability (2x)
years = df['Year'].unique()  # Get unique years
num_years = len(years)  # Get number of unique years
cols = 2  # Number of columns for subplots
rows = (num_years + cols - 1) // cols  # Calculate number of subplot rows
for i, year in enumerate(years, 1):  # Iterate over years
    plt.subplot(rows, cols, i)  # Create subplot
    sns.boxplot(x='Month_Name', y='Monthly beer production', data=df[df['Year'] == year], order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])  # Create boxplot for current year
    plt.title(f'Production Distribution for Year {year}', fontsize=18)  # Set subplot title (increased font size)
    plt.xlabel('Month', fontsize=16)  # Set x-axis label (increased font size)
    plt.ylabel('Monthly Beer Production', fontsize=16)  # Set y-axis label (increased font size)
    plt.xticks(fontsize=14, rotation=45)  # Set x-tick labels (increased font size)
    plt.yticks(fontsize=14)  # Set y-tick labels (increased font size)
plt.tight_layout()  # Adjust layout
plt.show()  # Show plot
```

```python
plt.figure(figsize=(24, 20))  # Set larger figure size for better readability (2x)
years = range(1990, 1996)  # Years from 1990 to 1995
num_years = len(years)  # Get number of years
cols = 2  # Number of columns for subplots
rows = (num_years + cols - 1) // cols  # Calculate number of subplot rows
for i, year in enumerate(years, 1):  # Iterate over years
    plt.subplot(rows, cols, i)  # Create subplot
    sns.boxplot(x='Month_Name', y='Monthly beer production', data=df[(df['Year'] == year) & (df['Year'] <= 1995)], order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])  # Create boxplot for current year within the specified range
    plt.title(f'Production Distribution for Year {year}', fontsize=18)  # Set subplot title (increased font size)
    plt.xlabel('Month', fontsize=16)  # Set x-axis label (increased font size)
    plt.ylabel('Monthly Beer Production', fontsize=16)  # Set y-axis label (increased font size)
    plt.xticks(fontsize=14, rotation=45)  # Set x-tick labels (increased font size)
    plt.yticks(fontsize=14)  # Set y-tick labels (increased font size)
plt.tight_layout()  # Adjust layout
plt.show()  # Show plot
```

```python
plt.figure(figsize=(24, 20))  # Set larger figure size for better readability (2x)
years = range(1990, 1996)  # Years from 1990 to 1995
num_years = len(years)  # Get number of years
cols = 2  # Number of columns for subplots
rows = (num_years + cols - 1) // cols  # Calculate number of subplot rows
for i, year in enumerate(years, 1):  # Iterate over years
    plt.subplot(rows, cols, i)  # Create subplot
    year_data = df[(df['Year'] == year) & (df['Year'] <= 1995)]
    plt.plot(year_data['Month'], year_data['Monthly beer production'], marker='o', linestyle='-')  # Line plot for monthly beer production
    plt.title(f'Monthly Beer Production for Year {year}', fontsize=18)  # Set subplot title (increased font size)
    plt.xlabel('Month', fontsize=16)  # Set x-axis label (increased font size)
    plt.ylabel('Monthly Beer Production', fontsize=16)  # Set y-axis label (increased font size)
    plt.xticks(fontsize=14, rotation=45)  # Set x-tick labels (increased font size)
    plt.yticks(fontsize=14)  # Set y-tick labels (increased font size)
plt.tight_layout()  # Adjust layout
plt.show()  # Show plot
```
